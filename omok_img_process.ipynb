```python
import cv2
import numpy as np
import time
```

```python
cap = cv2.VideoCapture(0)
```

```python
if (cap.isOpened() == False): 
    print("Unable to read camera feed")
```

```python
is_End = False
              
while(not is_End):
    is_Change = False
    
    while(not is_Change):
            frame_list = []
            
            is_Stable = False
            while(not is_Stable):
                ret, frame = cap.read()
                print(frame.shape)
                cv2.circle(frame, (90,65), 5, (255,0,0),-1)
                cv2.circle(frame, (570,75), 5, (255,0,0),-1)
                cv2.circle(frame, (0,230), 5, (255,0,0),-1)
                cv2.circle(frame, (640,240), 5, (255,0,0),-1)

                gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                pre_frame = gray_frame                

                if ret == True:
                    cv2.imshow('omok',gray_frame)
                    time.sleep(0.5)
                    
                    if cv2.waitKey(1) & 0xFF == ord('q'):
                        break
                        
                    if len(frame_list):
                        frame_list = np.append(frame_list,np.array([gray_frame]),axis=0)
                    else:
                        frame_list = np.array([gray_frame])
                    
                    if len(frame_list) == 5 :
                        if np.max(np.std(frame_list,axis=0)) < 50:
                            is_Stable = True
                            print('Stable!')
                        else:
                            print(np.max(np.std(frame_list,axis=0)))
                            frame_list = frame_list[1:]
                            print('Not Stable!')

                else:
                    break
            
            if np.mean(abs(pre_frame - frame_list[-1])) > 10:
                is_Change = True
                pre_frame = frame_list[-1]
                print('Change!')
            else:
                print('Not Change!')

    is_End = True
    print('End!')
    
cap.release()
cv2.destroyAllWindows() 
```
